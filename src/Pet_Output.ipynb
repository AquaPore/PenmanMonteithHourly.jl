```julia
using SolarPosition, Dates, CairoMakie

# define observer location (latitude, longitude, altitude in meters)
obs = Observer(52.35888, 1.0, 100.0)  # Van Gogh Museum, Amsterdam
times = collect(DateTime(2023, 1, 1, 0):Minute(60):DateTime(2023, 1, 1, 24));

positions = solar_position(obs, times, PSA(), HUGHES());
SolarNoon = SolarPosition.Utilities.next_solar_noon(obs,DateTime(2023, 1, 1), SPA())

positions_SolarNoon = solar_position(obs, SolarNoon, PSA(), HUGHES())


# plot elevation and azimuth over the day
fig = Figure(backgroundcolor = :transparent, textcolor= "#f5ab35", size = (800, 400))
ax1 = Axis(fig[1, 1], xlabel = "Time (hours)", ylabel = "Elevation (degrees)",
    title = "Solar Elevation on June 21, 2023 - Amsterdam",
    xticks = 0:3:24)
ax2 = Axis(fig[1, 2], xlabel = "Time (hours)", ylabel = "Azimuth (degrees)",
    title = "Solar Azimuth on June 21, 2023 - Amsterdam",
    xticks = 0:3:24)
ax3 = Axis(fig[1, 3], xlabel = "Time (hours)", ylabel = "Zenith (degrees)",
    title = "Solar Zenith on June 21, 2023 - Amsterdam", backgroundcolor = :transparent,
    xticks = 0:3:24)
times_hours = [Dates.hour(t) + Dates.minute(t)/60 for t in times]
lines!(ax1, times_hours, positions.elevation .- positions_SolarNoon.elevation, color = "#f5ab35")
lines!(ax2, times_hours, positions.azimuth.-positions_SolarNoon.azimuth, color = "#f5ab35")
lines!(ax3, times_hours, positions.zenith.-positions_SolarNoon.zenith, color = "#f5ab35")
fig
```
