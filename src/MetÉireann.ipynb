# Compiling the data of interest


```julia
SolarRadiation = []
Year           = []
Month          = []
Day            = []
Hour           = []
Minute         = []

for iT = 1:length(Months)
	Years[iT]
	Months[iT]

	if Months[iT] < 10
		Print_Zero = "0"
	else
		Print_Zero = ""
	end

	Folder = "mindata_$(StationNumber)_$(Years[iT])_$(Print_Zero)$(Months[iT])"
	File = "data_1min_$(StationNumber)_$(Years[iT])_$(Print_Zero)$(Months[iT]).csv"

	Path_Input = joinpath(Path_Output‚ÇÅ, Folder, File)
	println(Path_Input)
	@assert(isfile(Path_Input))

	Data‚ÇÄ      = CSV.File(Path_Input, header=true)

	soltot = convert(Vector{Float64}, Tables.getcolumn(Data‚ÇÄ,:soltot))
	year = convert(Vector{Int64}, Tables.getcolumn(Data‚ÇÄ,:year))
	month = convert(Vector{Int64}, Tables.getcolumn(Data‚ÇÄ,:month))
	day = convert(Vector{Int64}, Tables.getcolumn(Data‚ÇÄ,:day))
	hour = convert(Vector{Int64}, Tables.getcolumn(Data‚ÇÄ,:hour))
	minute = convert(Vector{Int64}, Tables.getcolumn(Data‚ÇÄ,:minute))

   SolarRadiation = cat(SolarRadiation, soltot ; dims=1)
   Year           = cat(Year,year ; dims=1)
   Month          = cat(Month, month ; dims=1)
   Day            = cat(Day, day; dims=1)
   Hour           = cat(Hour, hour; dims=1)
   Minute         = cat(Minute, minute; dims=1)
end

DayHour_Met = Dates.DateTimeMinute.(Year, Month, Day, Hour, Minute) #  <"standard"> "proleptic_gregorian" calendar

# Converting Solar radiation from [J/cm2/minute] ‚û° [Watts / m2/second]
SolarRadiation = SolarRadiation .* (1e4 / 60)
```

# Every 10 minutes data


```julia
ŒîT_Output = 10 * 60 ::Integer # Every 600 seconds
DayHour_Reduced = []

push!(DayHour_Reduced, DayHour_Met[1])

üéèBreak = false
iT = 2
while !(üéèBreak)
	if DayHour_Reduced[end] + Second(ŒîT_Output) > DayHour_Met[end]
		üéèBreak = true
		break
	else
		push!(DayHour_Reduced, DayHour_Reduced[end] + Second(ŒîT_Output))
		üéèBreak = false
		iT += 1
	end # if
end # while
```

```julia
SolarRadiation_Reduced = []
iTreduced = 1
for iT= 1:length(SolarRadiation)
	if DayHour_Reduced[min(iTreduced, length(DayHour_Reduced))] == DayHour_Met[iT]
		SolarRadiation_Reduced = append!(SolarRadiation_Reduced, SolarRadiation[iT])
		iTreduced += 1
	end
end

@assert length(SolarRadiation_Reduced)==length(DayHour_Reduced)
```

# Write to file

```julia
Path_Output = joinpath(Path_Output‚ÇÅ, "Roche point.csv")

Header = ["Date", "SolarRadiation[W/m¬≤]"]

CSV.write(Path_Output, Tables.table([DayHour_Reduced SolarRadiation_Reduced]), writeheader=true, header=Header, bom=true)
```
